# Libraries

# Read and display the image

# How is the image?

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))

howis(img)
```

# Selection of first channel (RED)

```python
X = img[:,:,0]
howis(X)
plt.imshow(X,cmap='gray')
plt.show()
```

# Thresholding - Simple Segmentation

```python
def segmenta(X,t):
  (N,M) = X.shape # toma el numero de filas y el numero de columnas
  Y = np.zeros((N,M)) #imagen de salida
  area = 0
  for i in range(N): #recorrer pixeles
    for j in range(M):
      if X[i,j] > t: # si es mayor que el umbral entonces que lo guarde
        Y[i,j] = 255
        area = area + 1 #contar el area de la imagen
  print('area = ',area)
  return Y #retorna lo valores a la imagen de salida
```

# Segmentation

```python
Y = segmenta(X,45)
howis(Y)
plt.imshow(Y)
plt.show()
```

# Definition of a homogeneous background

Each row is the original row minus its minimum


```python
(N,M) = X.shape
Xm    = np.zeros((N,M),np.uint8)  #organiza que el fondo sea mas homogeneo que tenga un mismo valor
for i in range(N):                #para que cuando segmentemos sea ams facil identificar lo que en realidad queremos
  xmin = np.min(X[i,:])
  Xm[i,:] = X[i,:] - xmin
plt.imshow(Xm,cmap='gray')
plt.show()
```

```python
Y = segmenta(Xm,38)
plt.imshow(Y,cmap='gray')
plt.show()
```
